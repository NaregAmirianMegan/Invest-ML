```python
#imports
import tensorflow as tf
import numpy as np
```

```python
#data
X = np.array([[0, 0], [1, 0], [0, 1], [1, 1]], dtype='float32')
Y = np.array([[0], [1], [1], [0]], dtype='float32')
```

```python
#constants
n_h1 = 3
n_h2 = 2
n_output = 1

training_steps = 5000
```

```python
#model
x = tf.placeholder(tf.float32, shape=[None, 2])
y = tf.placeholder(tf.float32, shape=None)
```

```python
#low level implementation 
with tf.variable_scope("low_level", reuse=tf.AUTO_REUSE):
    h1 = {'weights': tf.Variable(tf.random.normal([2, n_h1])), 
          'biases': tf.Variable(tf.random.normal([n_h1]))}

#     h2 = {'weights': tf.Variable(tf.random.normal([n_h1, n_h2])), 
#           'biases': tf.Variable(tf.random.normal([n_h2]))}

    output = {'weights': tf.Variable(tf.random.normal([n_h1, n_output])), 
              'biases': tf.Variable(tf.random.normal([n_output]))}

    l1 = tf.nn.relu(tf.add(tf.matmul(X, h1['weights']), h1['biases']))
#     l2 = tf.nn.relu(tf.add(tf.matmul(l1, h2['weights']), h2['biases']))
    out_l = tf.add(tf.matmul(l1, output['weights']), output['biases'])

    mse = tf.losses.mean_squared_error(y, out_l)
    loss = tf.reduce_mean(mse)
    optimizer = tf.train.AdamOptimizer(learning_rate=0.001)
    training_op = optimizer.minimize(loss)
    
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        for step in range(training_steps):
            _, curr_loss = sess.run([training_op, loss], feed_dict={x: X, y: Y})
            if step % 1000 == 0:
                print("Predictions:", np.round(sess.run(out_l, feed_dict={x: X}), decimals=0))
                print("Loss:", curr_loss)
        for var in tf.trainable_variables("low_level"):
            print(var)
```

```python
#mid level implementation
with tf.variable_scope("mid_level", reuse=tf.AUTO_REUSE):
    l1 = tf.layers.dense(x, n_h1, activation=tf.nn.relu)
    l2 = tf.layers.dense(l1, n_h2, activation=tf.nn.relu)
    out_l = tf.layers.dense(l2, n_output)

    mse = tf.losses.mean_squared_error(Y, out_l)
    loss = tf.reduce_mean(mse)
    optimizer = tf.train.AdamOptimizer(learning_rate=0.0001)
    training_op = optimizer.minimize(loss)
    
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        for step in range(training_steps):
            _, curr_loss = sess.run([training_op, loss], feed_dict={x: X, y: Y})
            if step % 1000 == 0:
                print("Predictions:", np.round(sess.run(out_l, feed_dict={x: X}), decimals=0))
                print("Loss:", curr_loss)
       # print(tf.trainable_variables('mid_level'))
```
